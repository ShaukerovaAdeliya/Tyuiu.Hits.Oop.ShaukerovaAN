**Описание задачи:**

Создать базовый класс Review в C#, который будет представлять отзывы о
продуктах или услугах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

**Требования к базовому классу Review:**

• **Атрибуты:** ID отзыва (ReviewId), Текст отзыва (Text), Рейтинг (Rating).

• **Методы:**

o DisplayReview(): метод для отображения отзыва.

o RateProduct(): метод для присвоения рейтинга продукту.

o GetReviewDetails(): метод для получения деталей отзыва.

**Требования к производным классам:**
1. ОтзывОбслуживания (ServiceReview): Должен содержать дополнительные
атрибуты, такие как Дата посещения (VisitDate).
Метод DisplayReview() должен быть переопределен для включения даты
посещения в отображение отзыва.
2. ОтзывТовара (ProductReview): Должен содержать дополнительные атрибуты,
такие как Идентификатор продукта (ProductId). Метод RateProduct() должен
быть переопределен для связывания рейтинга с конкретным продуктом.
3. ОтзывУслуги (ServiceReview) (если требуется третий класс): Должен
содержать дополнительные атрибуты, такие как Время начала услуги
(StartTime). Метод GetReviewDetails() должен быть переопределен для
отображения времени начала услуги вместе с другими деталями отзыва.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;
using System.Collections.Generic;

// Базовый класс для отзывов
public class Review
{
    public int ReviewId { get; set; }
    public string Text { get; set; }
    public int Rating { get; set; }
    public string Author { get; set; } // Автор отзыва
    public DateTime Date { get; set; } // Дата отзыва
    public string Category { get; set; } // Категория отзыва
    public List<string> Tags { get; set; } // Теги, связанные с отзывом

    public Review(int reviewId, string text, int rating, string author, DateTime date, string category)
    {
        ReviewId = reviewId;
        Text = text;
        Rating = rating;
        Author = author;
        Date = date;
        Category = category;
        Tags = new List<string>();
    }

    public virtual void DisplayReview()
    {
        Console.WriteLine($"Отзыв #{ReviewId}");
        Console.WriteLine($"Автор: {Author}");
        Console.WriteLine($"Дата: {Date}");
        Console.WriteLine($"Категория: {Category}");
        Console.WriteLine($"Текст: {Text}");
        Console.WriteLine($"Рейтинг: {Rating}");
        Console.WriteLine("Теги:");
        foreach (var tag in Tags)
        {
            Console.WriteLine($"- {tag}");
        }
        Console.WriteLine();
    }

    public virtual void RateProduct(int productId, int rating)
    {
        Rating = rating;
        Console.WriteLine($"Продукт #{productId} оценен на {Rating} звёзд.");
    }

    public virtual void GetReviewDetails()
    {
        Console.WriteLine($"Отзыв #{ReviewId}:");
        Console.WriteLine($"Текст: {Text}");
        Console.WriteLine($"Рейтинг: {Rating}");
    }

    public void AddTag(string tag)
    {
        Tags.Add(tag);
    }

    // Перегруженный метод для присвоения рейтинга с дополнительным комментарием
    public void RateProduct(int productId, int rating, string comment)
    {
        RateProduct(productId, rating);
        Console.WriteLine($"Комментарий: {comment}");
    }
}

// Производный класс для отзывов об обслуживании
public class ServiceReview : Review
{
    public DateTime VisitDate { get; set; }
    public string ServiceName { get; set; } // Название услуги

    public ServiceReview(int reviewId, string text, int rating, string author, DateTime date, string category, DateTime visitDate, string serviceName) : base(reviewId, text, rating, author, date, category)
    {
        VisitDate = visitDate;
        ServiceName = serviceName;
    }

    public override void DisplayReview()
    {
        base.DisplayReview();
        Console.WriteLine($"Дата посещения: {VisitDate}");
        Console.WriteLine($"Название услуги: {ServiceName}");
    }

    public void ScheduleNextVisit(DateTime nextVisitDate)
    {
        Console.WriteLine($"Запланировано следующее посещение на {nextVisitDate}");
    }
}

// Производный класс для отзывов о товарах
public class ProductReview : Review
{
    public int ProductId { get; set; }
    public string ProductName { get; set; } // Название товара

    public ProductReview(int reviewId, string text, int rating, string author, DateTime date, string category, int productId, string productName) : base(reviewId, text, rating, author, date, category)
    {
        ProductId = productId;
        ProductName = productName;
    }

    public override void RateProduct(int productId, int rating)
    {
        if (ProductId == productId)
        {
            base.RateProduct(productId, rating);
        }
        else
        {
            Console.WriteLine($"Ошибка: Неверный идентификатор продукта для рейтинга.");
        }
    }

    // Перегруженный метод для присвоения рейтинга с описанием продукта
    public void RateProduct(int productId, int rating, string productDescription)
    {
        if (ProductId == productId)
        {
            base.RateProduct(productId, rating);
            Console.WriteLine($"Описание продукта: {productDescription}");
        }
        else
        {
            Console.WriteLine($"Ошибка: Неверный идентификатор продукта для рейтинга.");
        }
    }
}

// Производный класс для отзывов об услугах
public class ServiceReview2 : Review
{
    public DateTime StartTime { get; set; }
    public string ServiceProvider { get; set; } // Исполнитель услуги

    public ServiceReview2(int reviewId, string text, int rating, string author, DateTime date, string category, DateTime startTime, string serviceProvider) : base(reviewId, text, rating, author, date, category)
    {
        StartTime = startTime;
        ServiceProvider = serviceProvider;
    }

    public override void GetReviewDetails()
    {
        base.GetReviewDetails();
        Console.WriteLine($"Время начала услуги: {StartTime}");
        Console.WriteLine($"Исполнитель услуги: {ServiceProvider}");
    }

    public void GetServiceDuration(DateTime endTime)
    {
        TimeSpan duration = endTime - StartTime;
        Console.WriteLine($"Продолжительность услуги: {duration}");
    }
}

// Generic класс для хранения списка отзывов
public class ReviewList<T> where T : Review
{
    private List<T> _reviews;

    public ReviewList()
    {
        _reviews = new List<T>();
    }

    public void AddReview(T review)
    {
        _reviews.Add(review);
    }

    public void DisplayAllReviews()
    {
        foreach (T review in _reviews)
        {
            review.DisplayReview();
        }
    }
}


// Создание объектов отзывов
ServiceReview serviceReview1 = new ServiceReview(1, "Отличное обслуживание!", 5, "Иван Иванов", DateTime.Now, "Ремонт", DateTime.Now.AddDays(-2), "Ремонт телефонов");
ProductReview productReview1 = new ProductReview(2, "Хороший товар!", 4, "Петр Петров", DateTime.Now.AddDays(-1), "Техника", 12345, "Смартфон");
ServiceReview2 serviceReview2 = new ServiceReview2(3, "Услуга была выполнена быстро.", 4, "Мария Сидорова", DateTime.Now, "Доставка", DateTime.Now.AddHours(-2), "Курьерская служба");

// Демонстрация полиморфизма
Review[] reviews = { serviceReview1, productReview1, serviceReview2 };
foreach (Review review in reviews)
{
    review.DisplayReview();
    review.RateProduct(12345, 3); // Переопределенный метод RateProduct()
    review.GetReviewDetails();
}

// Демонстрация generic класса
ReviewList<ServiceReview> serviceReviews = new ReviewList<ServiceReview>();
serviceReviews.AddReview(serviceReview1);
serviceReviews.AddReview(new ServiceReview(4, "Хорошая работа.", 4, "Анна Кузнецова", DateTime.Now, "Уборка", DateTime.Now.AddDays(-1), "Уборка квартир"));
serviceReviews.DisplayAllReviews();
```
